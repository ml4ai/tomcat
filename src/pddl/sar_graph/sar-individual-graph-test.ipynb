```python
paths = nx.shortest_path_length(G,source="O-L",target="L-E2",weight="dist")
```

```python
s_lengths = {}
for i in rooms:
    G.add_edges_from(rooms[i])
    s_lengths[i] = {}
    for j in rooms:
        if i != j:
            G.add_edges_from(rooms[j])
            p_length = nx.shortest_path_length(G,source=rooms[i][0][0],target=rooms[j][0][0],weight="dist")
            s_lengths[i][j] = p_length
            G.remove_edges_from(rooms[j])
    G.remove_edges_from(rooms[i])
```

```python
with open("../sar_shortest_path_lengths.json", "w") as json_file:
    json.dump(s_lengths,json_file,indent=4)
```
